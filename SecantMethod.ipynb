```python
def F(x):
    """x**5 - 2*x**4 - x**3 + +3*x**2 + x + 4"""
    return x**5 - 2*x**4 - x**3 + +3*x**2 + x + 4
```

```python
a = int(input("Enter a :"))
b = int(input("Enter b :"))
```

```python
def secant(a, b, M,e):
    error = []
    i = 0
    c = 0
    if F(a)  == F(b):
        print("Error")
        exit()
    else :
        while abs(F(b)) > e:
            if F(a)  == F(b):
                print("Error")
                return None
                
            c = b - ((b - a) * F(b)) / (F(b) - F(a))
            
            i += 1
            a,b = b, c
            if M<= i:
                return None
                
    return b
```

```python
root = secant(a, b, 10000, 0.000000005)
```
