```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def return_slope(x, y):
    return np.cosh(x)
```

```python
def eular_approximation(x0, y0, xn,n):
    h = (xn-x0)/n
    X, Y = [], []
    X.append(x0)
    Y.append(y0)
    x = x0
    y = y0
    for i in range(n):
        print(x, y)
        y += return_slope(x, y) * h
        x += h
        X.append(x)
        Y.append(y)
    return X, Y
```

```python
Ex, Ey = eular_approximation(0, 0, 6, 1000)
X = np.linspace(0, 6, 1001)
Y = [np.sinh(x) for x in X]
```

```python

plt.figure(figsize = (12, 8))
plt.plot(Ex, Ey, label = 'Approximated Result')
plt.plot(X, Y, linestyle = '--', label = 'Real Result')
n = 800
plt.vlines(Ex[n], 0, Ey[n], colors='gray', linestyles='--', label = f"y({Ex[n]:.2f}) = {Ey[n]:.2f}")
plt.title("Euler vs Exact Solution for dy/dx = cosh(x)")
plt.axhline(0)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
