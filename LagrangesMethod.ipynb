```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Lagrange(x, y, xp):
    sum = 0
    for i in range(len(x)):
        product = 1
        for j in range(len(x)):
            if i == j:
                continue
            product *= ((xp-x[j])/(x[i] - x[j]))
        sum += product * y[i]
    return sum
            
```

```python
p = -5
q = 5
```

```python
def adada(x):
    return np.sin(x/2) + np.cos(x/3) + 5
```

```python
X = np.arange(p, q)
Y = [adada(x) for x in X]
```

```python
A = np.linspace(p, q, 1000)
B = [Lagrange(X, Y, a) for a in A]
YY = [adada(x) for x in A]
```

```python
xp = -1
approx = Lagrange(X, Y, xp)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(A, YY, linestyle = "--",label = "Original Function")
plt.plot(A, B, label = "1/(1 + x**2)")
plt.vlines(xp, 0, approx, linestyle = '-.', color = 'grey', label = "sin(x/2) + cos(x/3) + 5")
plt.plot(xp, approx, marker = 'o', label = f"xp : {xp} , yp : {approx:.2f}")
plt.axhline(0)
plt.legend()
plt.title("Lagranges Approximation")

plt.grid()
```
