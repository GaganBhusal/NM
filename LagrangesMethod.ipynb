```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Lagrange(x, y, xp):
    sum = 0
    for i in range(len(x)):
        product = 1
        for j in range(len(x)):
            if i == j:
                continue
            product *= ((xp-x[j])/(x[i] - x[j]))
        sum += product * y[i]
    return sum
            
```

```python
p = -2
q = 2
```

```python
def f(x):
    return np.sin(2 * x)
```

```python
X = np.arange(p, q)
Y = [f(x) for x in X]
```

```python
A = np.linspace(p, q, 10000)
B = [Lagrange(X, Y, a) for a in A]
YY = [f(x) for x in A]
```

```python
xp = 0.5
approx = Lagrange(X, Y, xp)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(A, YY, linestyle = "--",label = "Original Function")
plt.plot(A, B, label = "sin(2x)")
plt.vlines(xp, 0, approx, linestyle = '-.', color = 'grey', )
plt.plot(xp, approx, marker = 'o', label = f"xp : {xp} , yp : {approx:.2f}")
plt.axhline(0)
plt.legend()
plt.title("Lagranges Approximation")

plt.grid()
```
