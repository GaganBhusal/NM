```python
import numpy as np
import matplotlib.pyplot as plt
```

Trapezoidal

```python
def f(x):
    return x ** 2
```

```python
def trap(a, b, n):
    h = (b-a)/n
    p = np.linspace(a+h, b-h, n-1)
    result = (f(a) + f(b) + 2 * sum(f(a) for a in p)) * h/2
    return result
```

```python
a, b, n = -2, 2, 10
print(f"The result is {trap(a, b, n)}")
```

```python
x = np.linspace(a, b, 1000)
y = f(x)

p = np.linspace(a, b, n)
pp = f(p)
```

```python
plt.plot(x, y)
plt.plot(p, pp, 'o')
plt.axhline(0)
plt.vlines(p, 0, pp, color = 'green', )

plt.fill_between(x, y, color='grey', alpha=0.25)
plt.title("Trapezoidal Rule")
plt.grid()
```

Simpsons 1/3

```python
def f(x):
    return x ** 2
```

```python
def simpsons13(a, b, n):
    m = np.array([1, 4, 1])
    h = (b-a)/n
    val = 0
    for i in range(0, n, 2):
        # print(a, i, h)
        x1 = np.linspace(a + i * h , a + (i + 2) * h , 3)
        y1 = [f(a) for a in x1]
        # print(x1, m)
        val += np.dot(y1, m) * h/3
    return val
```

```python
a, b, n = -2, 2, 100
print(f"The result is {simpsons13(a, b, n)}")
```

```python
x = np.linspace(a, b, 1000)
y = f(x)
plt.plot(x, y, color ='red')
plt.axhline(0)

plt.fill_between(x, y, color='grey', alpha=0.25)
plt.title("Simpsons1/3 Rule")
plt.grid()
```

```python
def simpsons38(a, b, n):
    m = np.array([1, 3, 3, 1])
    h = (b-a)/n
    val = 0
    for i in range(0, n, 3):
        # print(a, i, h)
        x1 = np.linspace(a + i * h , a + (i + 3) * h , 4)
        y1 = [f(a) for a in x1]
        # print(x1, m)
        val += np.dot(y1, m) * 3 *h/8
    return val
```

```python
a, b, n = -2, 2, 1000
print(f"The result is {simpsons38(a, b, n)}")
```

```python
x = np.linspace(a, b, 1000)
y = f(x)
plt.plot(x, y, color ='red')
plt.axhline(0)

plt.fill_between(x, y, color='grey', alpha=0.25)
plt.title("Simpsons 3/8 Rule")
plt.grid()
```

```python
def gauss2(a, b):

    w = np.array([1, 1])
    x = np.array([-1/np.sqrt(3), 1/np.sqrt(3)])
    p, q = (b - a)/2, (b + a)/2
    z = p * x + q
    return np.dot(w, f(z)) * p
    
```

```python
a, b = -2, 2
print(f"The result is {gauss2(a, b)}")
```

```python
x = np.linspace(a, b, 1000)
y = f(x)
plt.plot(x, y, color ='red')
plt.axhline(0)

plt.fill_between(x, y, color='grey', alpha=0.25)
plt.title("Gauss 2 Point Rule")
plt.grid()
```

```python
def gauss3(a, b):

    w = np.array([5/9, 8/9, 5/9])
    x = np.array([-np.sqrt(3/5), 0, np.sqrt(3/5)])
    p, q = (b - a)/2, (b + a)/2
    z = p * x + q
    return np.dot(w, f(z)) * p
    
```

```python
a, b = -2, 2
print(f"The result is {gauss3(a, b)}")
```
