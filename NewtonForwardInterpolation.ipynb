```python
import numpy as np
import matplotlib.pyplot as plt

def newton_fint(x,y,xp):
        n = len(x)
        D = np.zeros((n,n),dtype=float)

        D[:,0] = y

        for j in range(1,n):
            for i in range(n-j):
                  D[i,j] =  D[i+1,j-1] - D[i,j-1]

        print("Difference Table")

        for i in range(n):
            print (D[i,:n])

        h = x[1] - x[0]
        p = (xp-x[0])/h

        yp = y[0]
        prod = 1

        for j in range(1,n):
              prod *= (p-(j-1))/j 
              yp += prod*D[0,j]

        return yp, D
```

```python
x = np.arange(7, 13)
y = [a*2.9 + np.random.randint(-2, 2) for a in x]
xp = 9.5
# x_label = [f"x{i+1}" for i in range(3)]
```

```python
a, D = newton_fint(x, y, xp)
```
