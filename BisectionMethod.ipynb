<h2>Bisection Method</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
    return 2*x **3 + 7*x**2 + x + 2
```

```python
a = int(input("Enter x : "))
b = int(input("Enter y: "))
```

```python
def bisection(a, b, e):
    error = []
    i = 0
    if F(a) * F(b) >0:
        print("Error")
    else :
        while abs(a - b) > e:
            c = (a + b)/2
            if F(c) * F(a) < 0:
                b = c
            else:
                a = c
            print(a-b)
            error.append(a-b)
            i += 1
            max_iter = 2 * (np.log(b-a) - np.log(e)/np.log(2))
            if max_iter<= i:
                break
    return ((a + b)/2), error
```

```python
root, error = bisection(a, b, 0.000000005)
root
```

```python
plt.plot(range(len(error)), error, linestyle = '--')
plt.scatter(range(len(error)), error, marker = '.')
plt.grid()
```

```python
x = np.linspace(a, b, 100)
y = [F(i) for i in x]
```

```python
plt.plot(x, y, linestyle = '-', label = '2*x **3 + 7*x**2 + x + 2')
plt.axhline(0, linestyle = '-.')
plt.plot(root, F(root), 'o', label = f'a:{a}, b:{b}')
plt.axvline(root, linestyle = '-.')
plt.text(root+0.05, F(root)-1, f"({root:.2f},{F(root):.2f})")
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
