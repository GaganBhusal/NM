```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def F(x, y):
    return np.sinh(x)
```

```python
def RK4(x0, y0, xn, n):
    h = (xn-x0)/n
    x = []
    y = []
    x.append(x0)
    y.append(y0)
    for i in range(n):
        k1 = h * F(x[i], y[i])
        k2 = h * F(x[i] + h/2, y[i] + k1/2)
        k3 = h * F(x[i] + h/2, y[i] + k2/2)
        k4 = h * F(x[i] + h, y[i]+k3)

        k = (k1 + 2*k2 + 2*k3 + k4)/6

        x.append(x[i] + h)
        y.append(y[i] + k)
    return x, y
```

```python
X = np.linspace(0, 6, 1001)
Y = [np.cosh(x) for x in X]

RKx, RKy = RK4(0, 0, 6, 1000)
```

```python

plt.figure(figsize = (12, 8))
plt.plot(RKx, RKy, label = 'Approximated Result')
plt.plot(X, Y, linestyle = '--', label = 'Real Result')
plt.vlines(RKx[700], 0, RKy[700], colors='gray', linestyles='--', label = f"y({RKx[700]:.2f}) = {RKy[700]:.2f}")
plt.title("RK4 vs Exact Solution for dy/dx = sinh(x)")
plt.axhline(0)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
