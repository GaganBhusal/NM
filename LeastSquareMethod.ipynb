```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def return_summ(x, y):
    n = len(x)
    s_x = np.sum(x)
    s_y = np.sum(y)
    s_xx = np.sum([a**2 for a in x])
    s_yy = np.sum([b**2 for b in y])
    xy = np.sum(x * y)
    print(n, s_x, s_y, s_xx, s_yy, xy)
    return n, s_x, s_y, s_xx, s_yy, xy
```

```python
def LSM(x, y):
    n, s_x, s_y, s_xx, s_yy, xy = return_summ(x, y)
    D = n * s_xx - s_x * s_x
    D1 = s_y * s_xx - s_x * xy
    D2 = n * xy - s_x * s_y

    a = D1/D
    b = D2/D
    print(D1, D2, D)
    return a, b
    
```

```python
x = np.arange(10, 17)
y = [(np.random.randint(-3, 3) + 2 * a) for a in x]
```

```python
c, m = LSM(x, y)
X = np.linspace(10, 17, 100)
Y = [(c + m * b) for b in X]

YY = [(c + m * b) for b in x]
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(x, y, label = "Data Points")
plt.plot(X, Y,label = "Line")
plt.vlines(x, YY, y, linestyle = '--')
plt.grid()
plt.title("Least Square Method")
plt.legend()
```

```python
C, M = LSM(x, np.log(y))
```

```python
c = np.exp(C)
m = M
```

```python
X = np.linspace(0.1, 10, 100)
Y =[(c * np.exp(m * b)) for b in X]

YY = [c * np.exp(m * b) for b in x]
```
